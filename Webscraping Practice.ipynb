```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
import requests
import time
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
job_search = input('What job title are you searching for? ')
area = input('Where would you like your job to be located? ')
```

```python
url = f"https://www.indeed.com/jobs?q={job_search}&l={area}&vjk=ce23ed195a26b71e"
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
#Finding first box
jobBox = soup.find('div', class_='job_seen_beacon')

#Finding the job title 
jobTitleLocation = jobBox.find('h2', class_="jobTitle")
jobTitle = jobTitleLocation.find('span').text
print(jobTitle)

#Finding job location
CompanyName = jobBox.find('span', class_='companyName').text
print(CompanyName)

#Finding Location
CompanyLocation = jobBox.find('div', class_='companyLocation').text
print(CompanyLocation)

#Finding Salary if it exists
try:
    Salary = jobBox.find('div', class_='salary-snippet').text
    print(Salary)
except:
    print('No Salary Listed')
```

```python
#Looking through mutiple job listings:
#jobs = soup.find_all('div', class_="job_seen_beacon")
```

```python
job_title_list = []
company_name_list = []
company_location_list = []
salary_list = []
x = 0

for x in range(2):
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    jobs = soup.find_all('div', class_="job_seen_beacon")
    for result in jobs:

        # Identify and return title of listing
        jobTitleLocation = result.find('h2', class_="jobTitle")
        jobTitle = jobTitleLocation.find('span').text
        if jobTitle == 'new' :
            jobTitleLocation = result.find('h2', class_='jobTitle-newJob')
            jobTitles = jobTitleLocation.find_all('span')
            jobTitle = jobTitles[1].text
            #print(jobTitle)
        # Identify and return company name of listing
        CompanyName = result.find('span', class_='companyName').text
        # Identify and return the location of listing
        CompanyLocation = result.find('div', class_='companyLocation').text
        job_title_list.append(jobTitle)
        company_name_list.append(CompanyName)
        company_location_list.append(CompanyLocation)
        try:
            Salary = result.find('span', class_='salary-snippet').text
            salary_list.append(Salary)
        except AttributeError as e:
            salary_list.append('No Salary Information Provided')
    button_row = soup.find('ul', class_='pagination-list')
    all_buttons = button_row.find_all('li')
    next_button = all_buttons[-1]
    next_link = next_button.a['href']
    #print(next_link)
    browser.visit('https://www.indeed.com'+next_link)

        
items_dict = {
    'Job Title': job_title_list,
    'Company': company_name_list,
    'Location': company_location_list,
    'Salary': salary_list
}
```

```python
jobs_df = pd.DataFrame(items_dict)
jobs_df
```
