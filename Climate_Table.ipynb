```python
tables = pd.read_html(table_url)
```

```python
co2_info = tables[1]
co2_info = co2_info.rename(columns = {'Country[18]':'Country'})
co2_info
```

```python
co2_info.columns
#df1['Avg_Annual'] = df1['Avg_Annual'].str.replace(',', '')
co2_info[('Fossil CO2 emissions','2017(% of world)')] = co2_info[('Fossil CO2 emissions','2017(% of world)')].str.replace('%','')
co2_info
```

```python
co2_info[('Fossil CO2 emissions','2017(% of world)')] = co2_info[('Fossil CO2 emissions','2017(% of world)')].apply(lambda x: float(x))
#co2_info[('Fossil CO2 emissions','2017(% of world)')] = float(co2_info[('Fossil CO2 emissions','2017(% of world)')])
co2_info_sorted = co2_info.sort_values(by=[('Fossil CO2 emissions','2017(% of world)')], ascending=False)
co2_info_sorted
```
