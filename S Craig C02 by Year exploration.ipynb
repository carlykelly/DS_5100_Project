```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = [10, 6]
%config InlineBackend.figure_format = 'retina'

data = pd.read_csv('co2_renew_data.csv')
print(data.head())
print(len(data))
```

```python
df2 = data.copy()
df2= df2[['Country','1990','2005','2017']]

sns.set_style('whitegrid')
ax = sns.boxplot(y=df2['1990'])
#ax = sns.swarmplot(x="1990", data=df2, color=".25")

ax.set_title('CO2 Emmisions 1990')
plt.show()


df2['1990'].describe()
```

```python
ax1 = sns.boxplot(y='2005', data= df2)
ax.set_title('CO2 Emmisions 2005')
plt.show()

df2['2005'].describe()
```

```python
ax2 = sns.boxplot(y='2017', data= df2)
ax.set_title('CO2 Emmisions 2017')
plt.show()

df2['2017'].describe()
```

```python
fig = plt.figure()
x = np.arange(1)
ax = fig.add_axes([0,0,1,1])

ax.bar(x + 0, df2['1990'].mean(), color = 'b', width = 0.25)
ax.bar(x + 0.5, df2['2005'].mean(), color = 'b', width = 0.25)
ax.bar(x + 1, df2['2017'].mean(), color = 'b', width = 0.25)



ax.set_ylabel('Fossils C02 Emmisions in Megatons')
ax.set_title('Mean Fossils C02 Emmisions in Megatons by Year')
#ax.set_xticks(np.arange(3)+.25)
plt.xticks((0.05, .55, 1.05))
ax.set_xticklabels(['1990', '2005', '2017'])

#ax.set_yticks(np.arange(0, 81, 10))
#ax.legend(labels=['1990', '2005', '2017'])
plt.show()
```

```python
fig = plt.figure()
x = np.arange(1)
ax = fig.add_axes([0,0,1,1])

ax.bar(x + 0, df2['1990'].mean(), color = 'b', width = 0.25)
ax.bar( x+ .25, df2['1990'].median(), color = 'r', width = 0.25)

ax.bar(x + 0.75, df2['2005'].mean(), color = 'b', width = 0.25)
ax.bar(x+1, df2['2005'].median(), color = 'r', width = 0.25)

ax.bar(x + 1.5, df2['2017'].mean(), color = 'b', width = 0.25)
ax.bar(x+1.75, df2['2017'].median(), color = 'r', width = 0.25)


ax.set_ylabel('Fossils C02 Emmisions in Megatons',fontsize=16)
ax.set_title('Mean & Median Fossils C02 Emmisions in Megatons by Year', fontsize=20)
#ax.set_xticks(np.arange(3)+.25)
plt.xticks((0.12, .88, 1.63))
ax.set_xticklabels(['1990', '2005', '2017'])

#ax.set_yticks(np.arange(0, 81, 10))
ax.legend(labels=['Mean', 'Median'],fontsize=16)
plt.show()
```
