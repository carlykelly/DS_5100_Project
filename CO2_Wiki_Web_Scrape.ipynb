```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from urllib.request import urlopen
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_carbon_dioxide_emissions'
html = urlopen(url) 
soup = BeautifulSoup(html, 'html.parser')

tables = soup.find_all('table', {'class': 'wikitable sortable'})

country = []
C02_Em_1990 = []
C02_Em_2005 = []
C02_Em_2017 = []
C02_Em_PctWorld_2017 = []
C02_Em_2017vs1990change = []
Per_land_area = []
Per_capita = []

for table in tables:
    rows = table.find_all('tr')
    
    for row in rows[2:]:
        cells = row.find_all('td')
        
        
        if len(cells) > 1:
            col1 = cells[0]
            country.append(col1.text.strip())
            
            col2 = cells[1]
            C02_Em_1990.append(col2.text.strip())
            
            col3 = cells[2]
            C02_Em_2005.append(col3.text.strip())
            
            col4 = cells[3]
            C02_Em_2017.append(col4.text.strip())
            
            col5 = cells[4]
            C02_Em_PctWorld_2017.append(col5.text.strip())
            
            col6 = cells[5]
            C02_Em_2017vs1990change.append(col6.text.strip())
            
            col7 = cells[6]
            Per_land_area.append(col7.text.strip())
            
            col8 = cells[7]
            Per_capita.append(col8.text.strip())

df1 = pd.DataFrame(np.column_stack([C02_Em_1990, C02_Em_2005, C02_Em_2017, C02_Em_PctWorld_2017, C02_Em_2017vs1990change, Per_land_area, Per_capita]),index = country, columns = ('C02_Em_1990', 'C02_Em_2005', 'C02_Em_2017', 'C02_Em_PctWorld_2017', 'C02_Em_2017vs1990change', 'Per_land_area', 'Per_capita'))
df1
```
